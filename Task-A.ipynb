### A. Довольные аналитики. Условие задачи.

В Яндексе есть два аналитика — Ян и Петр. Каждый день они берутся за новую задачу. Ян получает удовольствие от работы, если сделал 2 задачи подряд успешно. А Петр — только когда после проваленной задачи в следующей добился успеха.

На сколько дней в среднем Яну ждать первого удовольствия дольше, чем Петру (успех и провал по каждой задаче равновероятны)?

Формат вывода
В ответе напишите число, округлив до двух знаков после запятой. Ответ может быть отрицательным.
Пример: -8.68

Пусть

__A__ - событие 01 (неудача, удача - Пётр), __B__ -  событие 11 (удача, удача - Ян)

### Решение с помощью теории вероятностей (матожидание).

#### Событие 11.

Пусть m11 - матожидание события 11. Тогда, после первых двух шагов, имеем 4 варианта с вероятностью по 1/4, и предстоит ещё ждать до события 11:

```
1/4, выпало 11, ещё ждать 0
1/4, выпало 00, ещё ждать m11
1/4, выпало 10, ещё ждать m11
1/4, выпало 01, ещё ждать n11 (для n позже запишем 2е уравнение)
1-ое уравнение: m11=2+1/4*0+1/4*m11+1/4*m11+1/4*n11
```

Для нахождения n11 выведем 2-ое уравнение:
```
1/2, выпало 1 и стало 11, ещё ждать 0
1/2, выпало 0 и стало 10, ещё ждать m11
2-ое уравнение: n11=1/2*1+1/2*(m11+1)
```
Решаем систему из 1-го и 2-го уравнений и получаем матожидание m11=6.

#### Событие 01.

Пусть m01 - матожидание события 01. Тогда, после первых двух шагов, имеем 4 варианта с вероятностью по 1/4, и предстоит ещё ждать до события 01:

```
1/4, выпало 11, ещё ждать m01
1/4, выпало 00, ещё ждать n01 (для n позже запишем 2е уравнение)
1/4, выпало 10, ещё ждать n01 (для n позже запишем 2е уравнение)
1/4, выпало 01, ещё ждать 0
1-ое уравнение: m01=2+1/4*m01+1/4*n01+1/4*n01+1/4*0
```

Для нахождения n01 выведем 2-ое уравнение:
```
1/2, выпадет 1 и станет 01, то есть ждать 1
1/2, выпадет 0 и станет 00, то есть ждать 1+m01
2-ое уравнение: n01=1/2*1+1/2*(1+m01)
```
Решаем систему из 1-го и 2-го уравнений и получаем матожидание m01=4.

__Ответ к задаче: m11-m01=6-4=2 (или 2.00)__

### Решение путём моделирования.

````python
import random


def test(n):
    """
    Функция получает длину строки и возвращает строку этой длины со случайными 0 и 1
    """
    s = ''
    for i in range(n):
        s += str(random.randint(0, 1))
    return s
````

````python
a = 0   # кумулятивный счётчик дней до события А (01)
b = 0   # кумулятивный счётчик дней до события В (11)
res = 0 # здесь будем хранить среднее значение
repeats = 2000000 # количество тестов, с его ростом среднее должно стабилизироваться
str_len = 75 # длина строки (на 1 больше количества дней, на котором тестируем)

for k in range(repeats):
    t = test(str_len) # запросили новую строку из 0 и 1
    a += t.find('01') # в конкретном тесте нашли дни до А и добавили в счётчик
    b += t.find('11') # в конкретном тесте нашли дни до В и добавили в счётчик
    res = (b - a) / (k + 1) # рассчитали среднее за все прошедшие тесты
    if k % 250 == 0:
        print(f'Прогресс: {round(100*k/repeats)}%, res: {round(res,5)}', end='\r')
        
res
````

Код в ячейке выше на каждые 250 итераций выдаёт текущее усреднённое значение "дней до 11 минус дней до 01". Количество тестов и дней подобраны экспериментально. Дальнейшее их увеличение сходимость не улучшает и лишь увеличивает время работы программы.

__Округлив, получаем ответ 2.00__
